## Koşullu İşlemler

```python
import pandas as pd
import seaborn as sns 
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df[df["age"] > 50].head()
```

```python
df[df["age"] > 50]["age"].count()
#Yaşı 50 den büyük olan kaç kişi olduğunu döndürüyor.
#ikinci "age" olmasaydı tüm variablelardaki 50 yaş üstünü döndüreckti.
```

```python
df.loc[df["age"] > 50, ["age","class"]].head()  
```

```python
df.loc[(df["age"] > 50) & (df["sex"]=="male"), ["age","class"]].head()  
#birden fazla koşul vererek seçim yaptık.
```

```python
df.loc[(df["age"] > 50) & (df["sex"]=="male") &(df["embark_town"]=="Cherbourg") , ["age","class","embark_town"]].head()  
```

```python
df.loc[(df["age"] > 50) & (df["sex"]=="male") &((df["embark_town"]=="Cherbourg") | (df["embark_town"]=="Southampton"))
       , ["age","class","embark_town"]].head()  
```

## Toplulaştıma ve Gruplama (Aggregation and Grouping)

```python
import pandas as pd
import seaborn as sns 
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df.groupby("sex")["age"].mean()
#Cinsiyetlerin yaş ortalaması.
```

```python
df.groupby("sex").agg({"age" : ["mean","sum"]})
```

```python
df.groupby("sex").agg({"age" : ["mean","sum"],
                       "survived":"mean"})
```

```python
df.groupby(["sex","embark_town"]).agg({"age" : ["mean","sum"],
                       "survived":"mean"})
```

```python
df.groupby(["sex","embark_town","class"]).agg({"age" : ["mean","sum"],
                       "survived":"mean"})
```

```python
df.groupby(["sex","embark_town","class"]).agg({"age" : ["mean","sum"],
                       "survived":"mean","sex" : "count"})
```

## Pivot Table

Pivot tablolar default olarak mean fonksiyonuyla gelir.

```python
import pandas as pd 
import seaborn as sns 
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df.pivot_table("survived","sex","embarked")
```

```python
df.pivot_table("survived","sex","embarked", aggfunc="std")
```

```python
df.pivot_table("survived","sex",["embarked","class"])
```

Yaş değişkeni kırılımına bakmamız için sayısal değişkeni kategorik değişkene çevirmemiz gerekmektedir.

Sayısal değişkeni kategorik değişkene çevirmek için cut veya qcut fonksiyonları kullanılır.Bunların arasındaki fark ise ; eğer değerleri neye göre ayıracağımızı biliyorsak cut (Örn. 0-10 yaş arasına çocuk , 10-25 yaş arasına genç diyebiliyorsak) eğer değerleri neye göre ayıracağımızı bilmiyorsak qcut (Çeyreklik değerlerine göre kendisi ayırır.)

```python
df["new_age"] = pd.cut(df["age"],[0,10,18,25,40,90])
```

```python
df.head()
```

```python
df.pivot_table("survived","sex","new_age")
```

```python
df.pivot_table("survived","sex",["new_age","class"])
```
