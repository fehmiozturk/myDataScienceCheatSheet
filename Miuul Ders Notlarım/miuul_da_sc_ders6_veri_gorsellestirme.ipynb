### Kategorik Değişkenler

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Marker

```python
y = np.array([13,28,11,100])
```

```python
plt.plot(y,marker = "h")
plt.show()
```

```python
markers  = ["o", '*', ".","," , 'x', 'X', '+', 'P' , 's', 'D', 'd', 'p', 'H', 'h']
```

## Line

```python
y = np.array([13,28,11,100])
plt.plot(y,linestyle="dashed")
plt.show()
```

```python
y = np.array([13,28,11,100])
plt.plot(y,linestyle="dotted")
plt.show()
```

```python
y = np.array([13,28,11,100])
plt.plot(y,linestyle="dashdot",color="r")
plt.show()
```

## Multiple Lines

```python
x = np.array([23,18,31,10])
y = np.array([13,28,11,100])
plt.plot(x)
plt.plot(y)
plt.show()
```

## Labels

```python
x =np.array([188, 85, 90, 95, 100, 105, 118, 115, 120, 125])
y =np.array([248, 258, 260, 270, 280, 298, 300, 310, 320, 330])
plt.plot(x, y)
# BAŞLIK
plt.title("First Line Plot")
# X ekseni 
plt.xlabel("X Ekseni Bilgilendirme")
#Y ekseni 
plt.ylabel("Y Ekseni Bilgilendirme")
# Arka Plan
plt.grid()
plt.show()
```

## Subplots

```python
# plot 1
x = np.array([88, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y =np.array([248, 250, 260, 276, 280, 290, 300, 310, 320, 330])
plt.subplot(1, 2, 1)
plt.title("1")
plt.plot(x,y)
plt.show()
# plot 2
x = np.array ([18, 8, 9, 9, 10, 15, 11, 15, 12, 15])
y = np.array([24, 20, 26, 27, 280, 29, 30, 30, 30, 30])
plt.subplot(1, 3, 2)
plt.title("2")
plt.plot(x, y)
plt.show()
# plot 3
x = np.array ([18, 8, 9, 9, 10, 15, 11, 15, 12, 15])
y = np.array([24, 20, 26, 27, 280, 29, 30, 30, 30, 30])
plt.subplot(1, 3, 3)
plt.title("3")
plt.plot(x, y)
plt.show()
```

# Seaborn

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

## Kategorik Değişkenler

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df["sex"].value_counts()
sns.countplot(x = df["sex"] , data=df)
plt.show()
```

## Sayısal Değişkenler

```python
sns.boxplot(x = df["total_bill"])
plt.show()
```
